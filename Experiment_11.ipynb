## Title: Categorical Data Analysis using Python


```python
import pandas as pd
data = {
    "Order ID": ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10'],

    'Category': ['Electronics', 'Clothing', 'Electronics', 'Grocery', 'Electronics', 'Clothing', 'Grocery', 'Clothing', 'Electronics', 'Grocery' ],

    'Payment_Method': ['Credit Card', 'PayPal', 'Credit Card', 'PayPal', 'Credit Card', 'PayPal','UPI', 'Cash', 'UPI', 'Cash' ],

    'Delivery_Status': ['Delivered', 'Pending', 'Delivered', 'Pending', 'Express',  'Delivered', 'Pending', 'Delivered', 'Pending', 'Express' ],

    'Customer_Type' : ['Regular', 'New', 'Regular', 'New', 'Returning', 'Regular', 'New', 'Regular', 'New', 'Returning'],
}

df = pd.DataFrame(data)
print(df)
```

```python
# Frequency Count
# Count no of orders in each Category

print(df['Category'].value_counts())

print(df['Payment_Method'].value_counts())
```

```python
# Unique Values
# Displays All Uniques Categories

print(df['Category'].unique())
print(df['Delivery_Status'].unique())
```

```python
# Number of Uniques Values

print(df['Category'].nunique())
```

```python
# Cross tabultion

#Category vs Payment

print(pd.crosstab(df['Category'], df['Payment_Method']))

```

```python
# Cross tabultion

#Delivery vs Customer Type

print(pd.crosstab(df['Delivery_Status'], df['Customer_Type']))
```

```python
# Percentage Distribution

print(df['Category'].value_counts(normalize=True)*100)
```

```python
# Filtering Data
# Example: Orders from Electronic Category

electronics_orders = df[df['Category'] == 'Electronics']
print(electronics_orders)
```

```python
# Grouiping
# Order BY Category and Payment method

print(df.groupby('Category')['Payment_Method'].value_counts())
```

```python
# Sorting data

Sorted_df = df.sort_values(by='Category')
print(Sorted_df)
```
